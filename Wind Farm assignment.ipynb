### Reading Individual_Wind_Forecasts csv file

```python
Individual_Wind_Forecasts = pd.read_csv("C:\\Users\\VASAVI\\Downloads\\Biourja take home assignment dataset.csv")
Individual_Wind_Forecasts
```

```python
def Zone(plant):
    if 'E' in plant:
        return 'East'
    elif 'N' in plant:
        return 'North'
    elif 'S' in plant:
        return 'South'
    else:
        return 'West'

Individual_Wind_Forecasts["Zone"] = Individual_Wind_Forecasts["Plant_Name"].apply(Zone)
Individual_Wind_Forecasts
```

### dividing into 4 different zones dataframes based on zones

```python
df = Individual_Wind_Forecasts.groupby('Zone')
for Zone,Zone_df in df:
    print(Zone)
    print(Zone_df)
```

```python
# Create a DataFrame to store the redistributed power output

redistributed_forecasts = Individual_Wind_Forecasts.copy()
```

```python
# Calculate the total individual forecast for each zone

zone_forecast_totals = {}
for zone in ['East', 'West', 'North', 'South']:
    zone_forecast_totals[zone] = Individual_Wind_Forecasts[Individual_Wind_Forecasts['Zone']== zone]['Forecast'].sum()
```

```python
zone_forecast_totals.items()
```

## Calculating ratio for each zone

### East Zonal Redistribution Ratio

```python
East_Zonal_Ratio = ZonalForecast[ZonalForecast['Zone'] == 'East']['Zonal_Forecast'] / zone_forecast_totals['East']
East_Zonal_Ratio
```

### West Zonal Redistribution Ratio

```python
West_Zonal_Ratio = ZonalForecast[ZonalForecast['Zone'] == 'West']['Zonal_Forecast'] / zone_forecast_totals['West']
West_Zonal_Ratio
```

### North Zonal Redistribution Ratio

```python
North_Zonal_Ratio = ZonalForecast[ZonalForecast['Zone'] == 'North']['Zonal_Forecast'] / zone_forecast_totals['North']
North_Zonal_Ratio
```

### South Zonal Redistribution Ratio

```python
South_Zonal_Ratio = ZonalForecast[ZonalForecast['Zone'] == 'South']['Zonal_Forecast'] / zone_forecast_totals['South']
South_Zonal_Ratio
```

### conactenating into a single series

```python
Zonal_Ratios = pd.concat([East_Zonal_Ratio,North_Zonal_Ratio,West_Zonal_Ratio,South_Zonal_Ratio])
Zonal_Ratios
```

### Redistributed output power for Each wind frame in East Zone

```python
df_East['Redistribution output'] = df_East['Forecast'] * Zonal_Ratios[0]
df_East.head()
```

### Redistributed output power for Each wind frame in West Zone

```python
df_West['Redistribution output'] = df_West['Forecast'] * Zonal_Ratios[2]
df_West.head()
```

### Redistributed output power for Each wind frame in North Zone

```python
df_North['Redistribution output'] = df_North['Forecast'] * Zonal_Ratios[1]
df_North.head()
```

### Redistributed output power for Each wind frame in South Zone

```python
df_South['Redistribution output'] = df_South['Forecast'] * Zonal_Ratios[3]
df_South.head()
```

### Concatenating all the datasets

```python
output = pd.concat([df_East,df_North,df_West,df_South])
output
```
